# import statements

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

# python code to read in data for all participants

```python
csv = os.listdir('./')
csv = [file for file in csv if '.csv' in file] 
```

# python code for helper functions (if necessary)

# python code to visualize accelerometer data for all participants and activities.

```python
dic = {}
i = 0
for file in csv:
    data = pd.read_csv(file,header=None)
    df = pd.DataFrame(data.values)
    df.columns=['sequential number','x acceleration','y acceleration','z acceleration','label']
    dic[i] = df
    i = i+1
    plt.plot(df['x acceleration'])
plt.show()
```

# 1a) python code to visualize pattern(s) of a specific activity.

```python
#Walking 
for x in range (15):
    label = (dic[x]['label']== 4)
    plt.plot(dic[x]['x acceleration'][label], label = 'x')
    plt.plot(dic[x]['y acceleration'][label], label = 'y')
    plt.plot(dic[x]['z acceleration'][label], label = 'z')
    
    plt.title('accelormeter data for walking')
    plt.legend()
    plt.xlabel('Ticks')
    plt.ylabel('Frequency')
    plt.show()
    
```

# 1b) Description of pattern(s) linked to activity.  Why would this activity produce this specific data pattern?

This activity produces high amounts of feqency data due to the nature of walking since you are moving up and down with each step and forward while you move and with A smaller but still impactful sway from side to side.
